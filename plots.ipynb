```python
import matplotlib.pyplot as plt

# Data for the first line (solid with two constant intervals and linear segments)
x1 = [1, 3, 4, 6, 7, 10]
# Data for the second line (dotted with two constant intervals and linear segments)
x2 = [1, 3, 4.3, 6, 7, 10]

y = [-1, -1, 1, 1, -1, -1]

# Plot the first line (solid)
plt.plot(x1, y, label='The network', color='blue', linestyle='-', marker='o')

# Plot the second line (dotted)
plt.plot(x2, y, label='Network with smaller norm', color='red', linestyle=':', marker='x')

# Adding titles and labels
plt.title('Dotted line has lower norm')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show legend
plt.legend(loc="upper right")

# Display the plot
plt.show()
```
